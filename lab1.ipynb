```python
#2
# Điểm đại học cao nhất và thấp nhất của 3 môn
pointDHMaxMin = df.aggregate({"DH1":["max","min"],
                            "DH2":["max","min"],
                            "DH3":["max","min"]})
print(pointDHMaxMin)
```

```python
# Điểm trung bình cao nhất, thấp nhất cửa từng môn qua từng lớp
pointTB1MaxMin = df.aggregate({"T1":["max","min"],"L1":["max","min"],
                                "H1":["max","min"],"S1":["max","min"],
                                "V1":["max","min"],"X1":["max","min"],
                                "D1":["max","min"],"N1":["max","min"]})
pointTB2MaxMin = df.aggregate({"T2":["max","min"],"L2":["max","min"],
                                "H2":["max","min"],"S2":["max","min"],
                                "V2":["max","min"],"X2":["max","min"],
                                "D2":["max","min"],"N2":["max","min"]})
pointTB3MaxMin = df.aggregate({"T3":["max","min"],"L3":["max","min"],
                                "H3":["max","min"],"S3":["max","min"],
                                "V3":["max","min"],"X3":["max","min"],
                                "D3":["max","min"],"N3":["max","min"]})
print(pointTB1MaxMin)
print(pointTB2MaxMin)
print(pointTB3MaxMin)
```

```python
# Đếm số lượng theo giới tính
StaGT = df.groupby("GT")["GT"].agg(["count"])
print(StaGT)
```

```python
# Đếm số lượng theo từng khu vực
StaKV = df.groupby("KV")["KV"].agg(["count"])
print(StaKV)
```

```python
# Đếm số lượng theo từng khối thi
StaKT = df.groupby("KT")["KT"].agg(["count"])
print(StaKT)
```

```python
#4
df["DT"].fillna(0, inplace =True)
```

```python
df["DT"]
```

```python
df["T1"].fillna(df["T1"].mean(), inplace =True)
```

```python
# Trung bình môn các năm 10 11 12
for i in ["1","2","3"]:
    df["TBM"+i] = (df["T"+i]*2 + df["L"+i]+ df["H"+i]+ df["S"+i]+
                   df["V"+i]*2 + df["X"+i]+ df["D"+i]+ df["N"+i])/10
for i in ["1","2","3"]:
    print("TBM"+i)
```

```python
for i in ["1","2","3"]:
    df.loc[df['TBM'+i] <= 5, 'XL'+i] = 'Y' 
    df.loc[(df['TBM'+i] <= 6.5) &(df['TBM'+i] > 5), 'XL'+i] = 'TB' 
    df.loc[(df['TBM'+i] < 8) &(df['TBM'+i] >= 6.5), 'XL'+i] = 'K' 
    df.loc[(df['TBM'+i] < 9) &(df['TBM'+i] >=8), 'XL'+i] = 'G' 
    df.loc[(df['TBM'+i] >= 9), 'XL'+i] = 'XS'
```

```python
df.head()
```

```python
#4
for i in ["1","2","3"]:
    df["US_TBM"+i] = df["TBM"+i]/10 *4
```

```python
df["US_TBM1"]
```

```python
#5
for i in df["KT"].unique():
    str(i)
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "A1" or i == "A") & (((df["DH1"]*2 + df["DH2"] + df["DH3"])/4) < 5), "KQXT"] = 0

    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) >= 5), "KQXT"] = 1
    df.loc[(i == "B") & (((df["DH1"] + df["DH2"]*2 + df["DH3"])/4) < 5), "KQXT"] = 0
    df.loc[(i not in ("A","A1","B")) & (((df["DH1"] + df["DH2"] + df["DH3"])/3) >= 5), "KQXT"] = 1
    df.loc[(i not in ("A","A1","B")) & (((df["DH1"] + df["DH2"] + df["DH3"])/3) < 5), "KQXT"] = 0
    #df.loc[(i != "A1" & i != "A" & i != "B") & (((df["DH1"] + df["DH2"] + df["DH3"])/3) >= 5), "KQXT"] = 1
    #df.loc[(i != "A1" & i != "A" & i != "B") & (((df["DH1"] + df["DH2"] + df["DH3"])/3) < 5), "KQXT"] = 0
```
